### Download the data

```python
url = "https://raw.githubusercontent.com/fedhere/PUI2016_fb55/master/Lab3_fb55/times.txt"

response = urllib2.urlopen(url)
data = response.readlines()

d = np.asarray([float(data[i]) for i in range(len(data))])
```

```python
d
```

### IDEA
The new Bus route is improving commute.

### Null Hypothesis:  
The commute time with the new bus route is on average the same as or longer than it was before.

$H_0 : Tnew \geqslant Told$
### Alternative Hypothesis:
The commute time with the new bus route is on average less than it was before.

$H_0 : Tnew < Told$

### In This example, we know about the old distribution:
$ Told \sim N(\mu = 36, \sigma = 6) $
 
### We choose the confident level as:
$ \alpha = 0.05 $

```python
mu = 36
sigma = 6
n = len(d)
```

```python
d.mean()
```

### Calculate the z score

```python
z = (mu - d.mean()) / sigma * sqrt(n)
z
```

As $2.5 > 2$, the new mean is farther than 2 sigmas away from the old mean 

$\Rightarrow$ We reject NULL Hypothesis
